# Import Pandas, Numpy and The ExcelSheet

```python
import pandas as pd
import numpy as np
df=pd.read_excel('covid19.xlsx')
df
```

# New column which Calc the percentage of HelthCare 
###   Convert The columns From object To float by excel and numeric(func)

```python
df['TotalDeaths']=pd.to_numeric(df['TotalDeaths'], errors='coerce')
df['HealthCare (%)'] =100- ((df['TotalDeaths']*100 ) / df['TotalCases'])
df
```

```python
df.dtypes
```

# clean the data and display all the dataframe

```python
df.dropna(inplace=True)
df.drop([221,222,223,224,225,228], inplace =True)
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
pd.options.display.max_rows=213
df
```

# Sort the DataFrame per (TotalCases)

```python
df=df.sort_values(by='TotalCases',ascending=False)
df
```

# replace israel with Palestine

```python
df['Country,Other'].replace({"Israel":"Palestine"},inplace=True)
df
```

```python
print('The total cases of all countries= ',df['TotalCases'].sum())
print('The total new cases of all countries= ',df['NewCases'].sum())
print('The total deaths of all countries= ',df['TotalDeaths'].sum())
```

```python
df_eg=df.groupby('Country,Other')
df_eg=df_eg.get_group('Egypt')
df_eg.plot(kind='bar',title='Egypt',figsize=(16,9))
```
